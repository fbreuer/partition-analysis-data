```python
import glob
import json
import pandas
pandas.set_option('display.notebook_repr_html', True)
from IPython.display import HTML
%pylab inline
from matplotlib import pyplot as plt
```

```python
def slurp_json(path):
    file = open(path,"r")
    r = json.load(file)
    file.close()
    return r
def spit_json(obj,path):
    file = open(path,"w")
    json.dump(obj,file)
    file.close()
```

```python
def show_benchmark_results(thepath,timeout=600.):
    data = [ slurp_json(path) for path in glob.glob(thepath) ]
    print "got %d files" % len(data)
    instances = list(set(( d["instance"] for d in data )))
    instances.sort()
    def lookup(predicate):
        return next( ( i for i in data if predicate(i) ), None )
    def get_total_time(instance,method):
        inst = lookup(lambda i: i["instance"] == instance and i["method"] == method)
        if inst == None:
            return None
        else:
            try:
                return inst["metrics"]["total_time"]
            except:
                return None
    def instance_data(instance):
        return { "instance": instance, 
            "polyhedral": get_total_time(instance,"1"),
            "omega": get_total_time(instance,"2"),
            "xin": get_total_time(instance,"3")}
    table = [ instance_data(inst) for inst in instances ] 
    df = pandas.DataFrame(table,columns=["instance","polyhedral","omega","xin"])
    df = df.replace({"-":None})
    print df
    html_result = HTML(df.to_html())
    df = df.fillna(timeout)
    for col in ["polyhedral","omega","xin"]:
        df[col] = df[col].astype('float64')
    def plot_runtimes(a,b):
        plt.figure(figsize=(8,8))
        plt.scatter(df[a].values,df[b].values)
        plt.xlabel(a)
        plt.ylabel(b)
        plt.yscale('log')
        plt.xscale('log')
    plot_runtimes("polyhedral","omega")
    plot_runtimes("polyhedral","xin")
    plot_runtimes("omega","xin")
    return html_result
```

```python
show_benchmark_results("2013-11-11-partitions/results/json/*.json",600.)
```
