```python
import glob
import json
```

```python
def slurp_json(path):
    file = open(path,"r")
    r = json.load(file)
    file.close()
    return r
def spit_json(obj,path):
    file = open(path,"w")
    json.dump(obj,file)
    file.close()
```

```python
data = [ slurp_json(path) for path in glob.glob("./results/json/*.json") ]
print "got %d files" % len(data)
data[0]
```

```python
data[171]
```

```python
instances = list(set(( d["instance"] for d in data )))
instances.sort()
def lookup(predicate):
    return next( ( i for i in data if predicate(i) ), None )
def get_total_time(instance,method):
    inst = lookup(lambda i: i["instance"] == instance and i["method"] == method)
    if inst == None:
        return None
    else:
        try:
            return inst["metrics"]["total_time"]
        except:
            return None
def instance_data(instance):
    return { "instance": instance, 
        "polyhedral": get_total_time(instance,"1"),
        "omega": get_total_time(instance,"2"),
        "xin": get_total_time(instance,"3")}
table = [ instance_data(inst) for inst in instances ] 
```

```python
import pandas
pandas.set_option('display.notebook_repr_html', True)
from IPython.display import HTML
```

```python
df = pandas.DataFrame(table,columns=["instance","polyhedral","omega","xin"])
df = df.replace({"-":None})
```

```python
HTML(df.to_html())
```

```python
df = df.fillna(300.)
for col in ["polyhedral","omega","xin"]:
    df[col] = df[col].astype('float64')
```

```python
%pylab inline
from matplotlib import pyplot as plt
```

Below we show running times of various instances as a scatter plot. Points above the diagonal mean that the algorithm on the vertical axis takes longer. Points below the diagonal mean that the algorithm on the horizontal axis takes longer. Points on the top edge mean that the vertical algorithm failes (but the horizontal one did not), whereas points on the right edge mean that the horizontal algorithm failed (but the vertical one did not). Point in the top right corner mean that both algorithms failed. Logarithmic scales are used throughout.

Ployhedral vs Xin. Polyhedral is faster in most cases, and it can also solve significantly more instances.

```python
plt.figure(figsize=(8,8))
plt.scatter(df["polyhedral"].values,df["xin"].values)
plt.xlabel('polyhedral')
plt.ylabel('xin')
plt.yscale('log')
plt.xscale('log')
```

Polyhedral vs. Omega. For most "easy" instances Omega is a bit faster. However Polyhedral is able to solve significantly more of the hard instances.

```python
plt.figure(figsize=(8,8))
plt.scatter(df["polyhedral"].values,df["omega"].values)
plt.xlabel('polyhedral')
plt.ylabel('omega')
plt.yscale('log')
plt.xscale('log')
```

Omega vs Xin. Omega is generally faster than Xin on easy instances. With hard instances, however, there is no clear winner. Omega is able to solve some instances that Xin can't and vice versa.

```python
plt.figure(figsize=(8,8))
plt.scatter(df["omega"].values,df["xin"].values)
plt.xlabel('omega')
plt.ylabel('xin')
plt.yscale('log')
plt.xscale('log')
```
